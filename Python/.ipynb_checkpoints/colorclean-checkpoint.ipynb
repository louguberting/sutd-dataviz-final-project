```python
import pandas as pd

rel_fol = "../RebrickableDec082025/"
#temp = "../temp/"

inventoryparts = pd.read_csv(rel_fol + "inventory_parts.csv", usecols=[0,1,2,3,4])
#temp = pd.read_csv(temp + "inventory_parts.csv", usecols=[0,1,2,3,4])
colors = pd.read_csv(rel_fol + "colors.csv", usecols=[0,1,2,3])
inventories = pd.read_csv(rel_fol + "inventories.csv", usecols=[0,2])
sets = pd.read_csv(rel_fol + "sets.csv", usecols=[0,2])

#print(inventoryparts)
#print(temp)

mergetable = inventoryparts.merge(colors, how='left', left_on='color_id', right_on='id')
mergetable = mergetable.merge(inventories, how='left', left_on='inventory_id', right_on='id')
mergetable = mergetable.merge(sets, how='left', on='set_num')

mergetable = mergetable[mergetable['year'] > 1940]
mergetable.sort_values(by='year')
 
#mergetable.to_csv('out.csv', index=False)
```

```python
mergetable['year'] = mergetable['year'].astype(int)
output = mergetable.groupby(['color_id', 'name', 'rgb', 'year'])['quantity'].sum().reset_index()
output.to_csv('colorrace.csv', index=False)
```
