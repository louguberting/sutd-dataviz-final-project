```python
import pandas as pd

themes_to_filter = [719, 760, 761]

rel_fol = "../RebrickableDec082025/"
sets = pd.read_csv(rel_fol + "sets.csv", usecols=[0,1,2,3])
print(sets)
themes = pd.read_csv(rel_fol + "themes.csv")
print(themes)

mergetable = sets.merge(themes, how='left', left_on='theme_id', right_on='id')
mergetable = mergetable.sort_values(by='year')
mergetable['group'] = mergetable['licensed'].apply(lambda x: 'Original' if x == 0 else 'Licensed')
mergetable['num'] = 0
mergetable = mergetable[~mergetable['theme_id'].isin(themes_to_filter)]
print(mergetable)
output = mergetable.groupby(['year', 'group'])['num'].count().reset_index()
print(output)
mergetable.to_csv('themedsets.csv', index=False)
output.to_csv('themeagg.csv', index=False)
```
