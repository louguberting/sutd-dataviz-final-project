```python
import pandas as pd

rel_fol = "../RebrickableDec082025/"
sets = pd.read_csv(rel_fol + "sets.csv", usecols=[0,2,3])
print(sets)
themes = pd.read_csv(rel_fol + "themes.csv")
print(themes)

mergetable = sets.merge(themes, how='left', left_on='theme_id', right_on='id')
mergetable.sort_values(by='year')
print(mergetable)
output = mergetable.groupby(['year', 'licensed'])
```
